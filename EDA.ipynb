```python
## Reading the dataset from the previously created csv-file in 'Dataset initialization (RQ1 & RQ2).ipynb'. 
import pandas as pd
df = pd.read_csv(r'/Users/noortje/Documents/880502-M-18 | Master Thesis/Python/Raw dataset for EDA.csv')

## Reviewing the loaded dataset
df
```

```python
## Further preparation of the data for EDA
## Obtaining the relevant features used for model construction, dropping irrelevant ones
df_eda = df[['area_id', 'f_beer', 'f_dairy', 'f_eggs', 'f_fats_oils', 'f_fish', 'f_fruit_veg', 'f_grains', 'f_meat_red', 
'f_poultry', 'f_readymade', 'f_sauces', 'f_soft_drinks', 'f_spirits', 'f_sweets', 'f_tea_coffee', 
'f_water', 'f_wine', 'male', 'income', 'education', 'distance_store', 'health_status']]
```

```python
## Exploration of summary statistics about the dataset
df_eda.describe()
```

```python
## Label Binarizing and obtaining the RQ1 dataset separately
import numpy as np
from sklearn.preprocessing import LabelBinarizer

## Binarizing the target feature in 0 and 1
label = LabelBinarizer()
target_bin = label.fit_transform(df_eda['health_status'])

## Putting it in a DataFrame
df_rq1 = pd.DataFrame(target_bin, columns = ['health_status'])

## Concatenating the numeric features with the target
df_rq1 = pd.concat([df_eda[['male', 'income', 'education', 'distance_store']], df_rq1], axis=1)

## Reviewing the operation
df_rq1
```

```python
## Label Binarizing and obtaining the RQ2 dataset separately
import numpy as np
from sklearn.preprocessing import LabelBinarizer

## Binarizing the target feature in 0 and 1
label = LabelBinarizer()
target_bin = label.fit_transform(df_eda['health_status'])

## Putting it in a DataFrame
df_rq2 = pd.DataFrame(target_bin, columns = ['health_status'])

## Concatenating the numeric features with the targett
df_rq2 = pd.concat([df_eda[['f_beer', 'f_dairy', 'f_eggs', 'f_fats_oils', 'f_fish', 'f_fruit_veg', 'f_grains', 'f_meat_red', 
'f_poultry', 'f_readymade', 'f_sauces', 'f_soft_drinks', 'f_spirits', 'f_sweets', 'f_tea_coffee', 
'f_water', 'f_wine']], df_rq2], axis=1)

## Reviewing the operation
df_rq2
```

```python
## Dropping the duplicates for RQ1
df_rq1 = df_rq1.drop_duplicates()

## Reviewing the operation
df_rq1
```

```python
## Checking missing values for RQ1
total_missing_values = df_rq1.isnull().sum().sum()

## Reviewing the missing values
print(total_missing_values)
```

```python
## Dropping the columns with missing values for RQ1
df_rq1 = df_rq1.dropna()

## Reviewing the operation
df_rq1
```

```python
## Exploratory Data Analysis
## Checking missing values for RQ2
total_missing_values = df_rq2.isnull().sum().sum()

## Reviewing the missing values
print(total_missing_values)
```

```python
## Checking histograms for data distributions
# NOTE: This code block was re-used for every feature, i.e., 'feature' variable replaced for every feature
import matplotlib.pyplot as plt

## Defining the feature in question
feature = df_rq1['income']

## Obtaining the histogram
fig, ax = plt.subplots(figsize =(5, 4))
ax.hist(feature, bins = 100)
ax.set_ylabel('Number of observations', fontsize = 10, fontweight = 'bold')
ax.set_xlabel('Number of income deprivated people', fontsize = 10, fontweight = 'bold')
plt.show()
```

```python
## Checking Q-Q plots for data distributions
# NOTE: This code block was re-used for every feature, i.e., 'feature' variable replaced for every feature
import statsmodels.api as sm
import pylab as py
from scipy import stats

## Obtaining the Q-Q plot
fig, ax = py.subplots(figsize =(5, 4))
sm.qqplot(feature, fit = True, line = '45', ax=ax, dist = stats.norm)
ax.set_xlabel('Theoretical Quantiles', fontsize = 10, fontweight = 'bold')
ax.set_ylabel('Sample Quantiles', fontsize = 10, fontweight = 'bold')
py.show()
```

```python
## Checking boxplots for outlier detection
# NOTE: This code block was re-used for every feature, i.e., 'feature' variable replaced for every feature

## Setting the variabes
feature = df_rq1[['income']]

## Obtaining the boxplots
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize = (5, 4))
ax.set_ylabel('Range of values', fontsize = 10, fontweight = 'bold')
feature.plot.box(ax = ax)
plt.show()
```

```python
## Correlation analysis for RQ1 dataset
import seaborn as sns
import matplotlib.pyplot as plt

## Create a dataframe with only numeric input features
df_rq1_corr = df_rq1[['male', 'income', 'education', 'distance_store']]

## Obtaining the correlation heatmap
corr = df_rq1_corr.corr(method = 'spearman')
plt.figure(figsize = (5, 4))
sns.heatmap(corr, cmap = 'coolwarm', annot = True, cbar = False)
plt.show()
```

```python
## Correlation analysis for RQ2 dataset
import seaborn as sns
import matplotlib.pyplot as plt

## Create a dataframe with only numeric input features
df_rq2_corr = df_rq2[['f_beer', 'f_dairy', 'f_eggs', 'f_fats_oils', 'f_fish', 'f_fruit_veg', 'f_grains', 'f_meat_red', 
'f_poultry', 'f_readymade', 'f_sauces', 'f_soft_drinks', 'f_spirits', 'f_sweets', 'f_tea_coffee', 
'f_water', 'f_wine']]

## Obtaining the correlation heatmap
corr2 = df_rq2_corr.corr(method = 'spearman')
plt.figure(figsize = (13, 13))
sns.heatmap(corr2, cmap = 'coolwarm', annot = True, cbar = False)
plt.show()
```

```python
## Dropping the correlated features for RQ1 and RQ2
df_rq1 = df_rq1[['male', 'education', 'distance_store', 'health_status']]

df_rq2 = df_rq2[['f_beer', 'f_dairy', 'f_eggs', 'f_fats_oils', 'f_meat_red',
'f_sauces', 'f_spirits', 'f_sweets', 'f_tea_coffee', 'f_water', 'f_wine', 'health_status']]
```

```python
## PointBiserialR Correlation Coefficient
## See manuscript for results
from scipy.stats import pointbiserialr

r, p = pointbiserialr(df_rq1['health_status'], df_rq1['income'])
print(f"Point-Biserial Correlation Coefficient: {r:.3f}")
print(f"P-value: {p:.3f}")
```
