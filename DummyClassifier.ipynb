```python
## DUMMYCLASSIFIER FOR RQ1 DATASET
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score, balanced_accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import RobustScaler
import numpy as np

## Assign data to X and Y
X = df1.drop('health_status', axis=1).values
Y = df1['health_status'].values

## Splitting the data into a training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, Y, 
test_size = 0.23, stratify = Y)

## Feature scaling using RobustScaler to put all features to the same scale
scaler = RobustScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

## Implementing SMOTE on training data, to train the model on a balanced dataset
smote = SMOTE(random_state=101, sampling_strategy = 'minority')
X_res, y_res = smote.fit_resample(X_train, y_train)

## Assigning sample weights
weights = np.zeros(len(y_res))
weights[y_res == 1] = 0.991
weights[y_res == 0] = 0.009

## Defining and fitting the model
model = DummyClassifier(random_state = 101, strategy = 'most_frequent')
model.fit(X_res, y_res)

## Making predictions for the dummy on testing data
pred = model.predict(X_test)

## Evaluating the baseline score
print('Baseline Recall-score (positive class) =', recall_score(y_test, pred, pos_label = 1))
print('Baseline Balanced Accuracy Score =', balanced_accuracy_score(y_test, pred))
```
