```python
## Combining all 12 separate .csv files to one big dataset
## Source of code: https://stackoverflow.com/questions/20906474/import-multiple-csv-files-into-pandas-and-concatenate-into-one-dataframe
import glob
import os
import pandas as pd

data = pd.concat(map(pd.read_csv, glob.glob(os.path.join('/Users/noortje/Documents/880502-M-18 | Master Thesis/Data/Dataset/LSOA/*.csv'))))

## Reviewing concatenation
data
```

```python
## Dropping irrelevant features
df = data[['area_id', 'weight', 'energy_tot', 'fat', 'saturate', 'salt','sugar','protein','carb','fibre','alcohol','f_beer',
'f_dairy', 'f_eggs', 'f_fats_oils', 'f_fish', 'f_fruit_veg', 'f_grains', 'f_meat_red', 'f_poultry', 'f_readymade', 'f_sauces',
'f_soft_drinks', 'f_spirits', 'f_sweets', 'f_tea_coffee', 'f_water', 'f_wine', 'male', 'female', 'age_0_17', 'age_18_64', 'age_65+',
'avg_age']]

## Reviewing the operation
df
```

```python
## Adding the additional demographic characteristics
## Demographic 1: Income
## Source of data: https://www.gov.uk/government/statistics/english-indices-of-deprivation-2015
income = pd.read_excel(r'/Users/noortje/Documents/880502-M-18 | Master Thesis/Data/Demographics datasets/File8_Income_2.xlsx')

df2 = pd.merge(df, income, on='area_id')
```

```python
## Adding the additional demographic characteristics
## Demographic 2: Education level
## Source of data: https://www.gov.uk/government/statistics/english-indices-of-deprivation-2015
education = pd.read_excel(r'/Users/noortje/Documents/880502-M-18 | Master Thesis/Data/Demographics datasets/File8_adult_skill.xlsx')

df3 = pd.merge(df2, education, on='area_id')
```

```python
## Demographic 3: Distance towards stores
## Source of data: https://www.gov.uk/government/statistics/english-indices-of-deprivation-2015
distance_store = pd.read_excel(r'/Users/noortje/Documents/880502-M-18 | Master Thesis/Data/Demographics datasets/File8_distance_store.xlsx')

df4 = pd.merge(df3, distance_store, on='area_id')
```

```python
## Preparing the 'nutrient' features for the Nutri-Score calculation
## Setting the weight to 100g and perform the associated calculations on the DataFrame

## Renaming the dataframe for simplicity
df_calc = df4

## Performing operations
df_calc['energy_tot'] = ((100 * df_calc['energy_tot']) / df_calc['weight'])
df_calc['energy_tot_KJ'] = (df_calc['energy_tot'] / 4.1868) ## must be in (kJ)
df_calc['fat'] = ((100 * df_calc['fat']) / df_calc['weight'])
df_calc['saturate'] = ((100 * df_calc['saturate']) / df_calc['weight'])
df_calc['salt'] = ((100 * df_calc['salt']) / df_calc['weight']) ## must be in (mg)
df_calc['salt_MG'] = ((df_calc['salt'] / 2.5) * 1000) # Should be divided by 2.5 according to guidelines
df_calc['sugar'] = ((100 * df_calc['sugar']) / df_calc['weight'])
df_calc['protein'] = ((100 * df_calc['protein']) / df_calc['weight'])
df_calc['fibre'] = ((100 * df_calc['fibre']) / df_calc['weight'])
df_calc['weight'] = 100

## Renaming the dataframe for simplicity
df = df_calc

## Reviewing the operation
df
```

```python
## Constructing the Nutri-Score for each instance
## Source of the library: https://pypi.org/project/pyNutriScore/
## Library GitHub page: https://github.com/lemonhead94/pyNutriScore/blob/main/pyNutriScore/interfaces/scoreTable.json 
from pyNutriScore import NutriScore
import numpy as np

value = []

for i, row in df.iterrows():
    result = NutriScore().calculate_class(
    {
    'energy': df['energy_tot_KJ'].values[i],
    'fibers': df['fibre'].values[i],
    'fruit_percentage': 0, ## Set to 0, "modfied" version
    'proteins': df['protein'].values[i],
    'saturated_fats': df['saturate'].values[i],
    'sodium': df['salt_MG'].values[i],
    'sugar': df['sugar'].values[i],
    }, 'solid')
    value.append(result)

df_nutri = {'Nutri-Score':value}
df2 = pd.DataFrame(df_nutri)
```

```python
## Reviewing the distribution of the Nutri-Score labels through the dataset
print('A =', df2.value_counts()['A'])
print('B =', df2.value_counts()['B'])
print('C =', df2.value_counts()['C'])
## "D" has no value since it is not present in the DataFrame
print('E =', df2.value_counts()['E'])
```

```python
## Assigning the binary classification label to the dataset, according to obtained Nutri-Score label
df.loc[df2['Nutri-Score'] == 'A', 'health_status'] = 'Healthy'
df.loc[df2['Nutri-Score'] == 'B', 'health_status'] = 'Healthy'
df.loc[df2['Nutri-Score'] == 'C', 'health_status'] = 'Unhealthy'
df.loc[df2['Nutri-Score'] == 'D', 'health_status'] = 'Unhealthy'
df.loc[df2['Nutri-Score'] == 'E', 'health_status'] = 'Unhealthy'

## Reviewing the distribution of the classification label 
print('Healthy =', df['health_status'].value_counts()['Healthy'])
print('Unhealthy =', df['health_status'].value_counts()['Unhealthy'])
```

```python
## The dataset is now constructed and ready to use for EDA. 
## Saving this dataset into a .csv file for further use within the thesis project
df.to_csv('Raw dataset for EDA.csv', index = False)
```
